```python
from pyspark.sql.functions import col, count

def check_not_empty(df, table):
    cnt = df.count()
    if cnt == 0:
        raise Exception(f"{table} has 0 rows")
    return cnt


def check_null_rate(df, table, column_name, max_pct=0.05):
    total = df.count()
    nulls = df.filter(col(column_name).isNull()).count()
    pct = nulls / total if total else 0

    if pct > max_pct:
        raise Exception(f"{table}.{column_name} null rate too high: {pct:.2%}")

    return pct


def check_duplicates(df, key_cols):
    dupes = (
        df.groupBy(key_cols)
          .count()
          .filter("count > 1")
          .count()
    )

    if dupes > 0:
        raise Exception(f"Duplicate keys found: {dupes}")
```
