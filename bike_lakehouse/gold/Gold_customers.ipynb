## Business transformation

```python
log_event("gold", "dim_customers", "START", "Building dimension")
```

```python
query = """
SELECT
    ROW_NUMBER() OVER (ORDER BY ci.customer_id) AS customer_key,
    ci.customer_id,
    ci.customer_number,
    ci.first_name,
    ci.last_name,
    la.country,
    ci.marital_status,
    CASE
        WHEN ci.gender <> 'n/a' THEN ci.gender
        ELSE COALESCE(ca.gender, 'n/a')
    END AS gender,
    ca.birth_date AS birthdate,
    ci.created_date AS create_date
FROM silver.crm_customers ci
LEFT JOIN silver.erp_customers ca
    ON ci.customer_number = ca.customer_number
LEFT JOIN silver.erp_customer_location la
    ON ci.customer_number = la.customer_number
"""
df = spark.sql(query)
```

```python
df.write.mode("overwrite").option("overwriteSchema", "true").format("delta").saveAsTable("workspace.gold.dim_customers")
row_count = check_not_empty(df, "gold.dim_customers")
log_event("gold", "dim_customers", "SUCCESS", "Write complete", rows=row_count)
```

```python
check_null_rate(df, "gold.dim_customers", "customer_key")
check_null_rate(df, "gold.dim_customers", "customer_id")
check_duplicates(df, ["customer_key"])
check_duplicates(df, ["customer_id"])
```
